#Start variáveis

```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
DT = 0.01 #Intervalo de tempo
y0=0 #Posicao inicial
v0=100 #velocidade inicial
g=-10 #gravidade
x0 = 0 #Posicao inicial
t0= 0 #tempo inicial


tMax = 50 #Tempo máximo
theta = 45 #Angulo de lancamento
vel=[v0]
vx=[v0*math.cos(math.radians(theta))] #Velocidade na direcao x
vy=[v0*math.sin(math.radians(theta))] #velocidade na direcao y
y = [y0] #Posicao inicial do eixo y
x = [x0] #Posicao inicial do eixo x
k = 0.1 #Valor do atrito
m = 1 #Massa do objeto
tempo=np.arange(0,tMax,DT)

Ec = [(m*v0**2)/2] #Energia cinetica
Ecx = [(m*vx[0]**2)/2] #Energia cinetica eixo x
Ecy = [(m*vy[0]**2)/2] #Energia cinetica eixo y

Edx = [m*g - k*vx[0]] #Energia dissipada eixo x
Edy = [m*g - k*vy[0]] #Energia dissipada eixo y

Ep = [m*(-g)*v0]      #Energia Potencial 
Epx = [m*(-g)*vx[0]]  #Energia Potencial x
Epy = [m*(-g)*vy[0]] #Energia Potencial y


Etx = [Ecx[0]+Epx[0]] #Energia total x
Ety = [Ecy[0]+Epy[0]] #Energia total y
```

#Cálculos

```python
for i in np.arange(DT,tMax,DT):
  DistanciaX=-k/m*vx[int((i-0.01)*100)] #Obtendo distancia no eixo 'x'
  DistanciaY=(g-k/m*vy[int((i-0.01)*100)]) #Obtendo a distancia no eixo 'y'
  #Calculando eixo x (Distancia)
  Dvx = (DistanciaX)*DT
  Dx = vx[int((i-0.01)*100)]*DT
  #print(Dx)

  #Atualizando eixo x na lista
  x.append(x[int((i-0.01)*100)]+Dx)
  vx.append(vx[int((i-0.01)*100)]+Dvx)

  #Calculando eixo y (Altura do objeto)
  Dvy = (DistanciaY)*DT
  Dy = vy[int((i-0.01)*100)]*DT

  #Atualizando eixo y na lista
  if (y[int((i-0.01)*100)]+Dy) <= 0:
    y.append(0)
    vy.append(-(vy[int((i-0.01)*100)]+Dy))
  else:
    y.append(y[int((i-0.01)*100)]+Dy)
    vy.append(vy[int((i-0.01)*100)]+Dvy)

  #Calculando a energia cinetica no eixo x
  Ecx.append(Ecx[int((i-0.01)*100)]+Dx)
  Ecy.append(Ecy[int((i-0.01)*100)]+Dy)
  
  Ec.append(Ecx[int((i-0.01)*100)] + Ecy[int((i-0.01)*100)])
  
  #Calculo da velocidade resultante
  vel.append(math.sqrt(vy[int(i*100)]**2+vx[int(i*100)]**2))
  #print(Ecx)

  #Calculo da energia dissipada
  Edx.append(m*g - k*vx[int((i-0.01)*100)])
  Edy.append(m*g - k*vy[int((i-0.01)*100)])

  #Calculo da Energia Potencial
  Epx.append(m*(-g)*vx[int((i-0.01)*100)])  #Energia Potencial x
  Epy.append(m*(-g)*vy[int((i-0.01)*100)]) #Energia Potencial y

  #Calculo da energia total
  Etx.append(Ecx[int((i-0.01)*100)] + Epx[int((i-0.01)*100)])
  Ety.append(Ecy[int((i-0.01)*100)] + Epy[int((i-0.01)*100)])



plt.plot(x, y)

plt.xlabel('x - Distancia')
plt.ylabel('y - Altura')

plt.title("Projétil")
plt.show()
```

No gráfico acima temos a altura da partícula em relação a distância percorrida. Analisando o gráfico percebemos que o movimento do projétil é de um objeto que quica, pois ele bate no chão e volta. Numa altura inferior a anterior pois o atrito do ar o impede de chegar a mesma altura anteriormente.
